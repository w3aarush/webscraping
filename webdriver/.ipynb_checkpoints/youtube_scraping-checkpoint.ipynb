```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
```

```python
options.add_argument('--head')
```

```python
browser = webdriver.Chrome(options=options)
```

```python
link = 'https://www.youtube.com/@GeeksforGeeksVideos/videos'
```

```python
browser.get(link)
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
videos = soup.find_all('ytd-rich-item-renderer',class_='style-scope ytd-rich-grid-renderer')
```

```python
videos[0].find_all('a',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media")[0].get('title')
```

```python
videos[0].find_all('a',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media")[0].get('href')
```

```python
videos[0].find_all('div',class_='yt-badge-shape__text')[0].get_text()
```

```python
videos[0].find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')[0].text
```

```python
videos[0].find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')[1].text
```

```python
video_list = []
for video in videos:
    title = video.find_all('a',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media")[0].get('title')
    link = 'https://www.youtube.com/'+video.find_all('a',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media")[0].get('href')
    duration = video.find_all('div',class_='yt-badge-shape__text')[0].get_text()
    views = video.find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')[0].text
    upload_time = video.find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')[1].text
    video_list.append([title, link, duration, views, upload_time])
```
