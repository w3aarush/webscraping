```python
import pandas as pd
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('./book_data_part1.csv')
```

```python
df.head()
```

```python
df.book_link.head()
```

```python
data = []

for link in tqdm(df['book_link']):
    res     = requests.get(link)
    soup    = BeautifulSoup(res.text, 'html.parser')

    '''
    typ     = soup.find('ul', class_='breadcrumb').find_all('a')[-1].text
    upc     = soup.find('table', class_ = 'table table-striped').find_all('td')[0].text
    price_x = soup.find('table', class_ = 'table table-striped').find_all('td')[2].text[2:]
    price_i = soup.find('table', class_ = 'table table-striped').find_all('td')[3].text[2:]
    tax     = soup.find('table', class_ = 'table table-striped').find_all('td')[4].text[2:]
    qn   = soup.find('table', class_ = 'table table-striped').find_all('td')[5].text
    reviews = soup.find('table', class_ = 'table table-striped').find_all('td')[6].text
    description = soup.find('article', class_ = 'product_page').find_all('p')[-1].text
    '''
    typ   = soup.find('ul', class_='breadcrumb').find_all('a')[-1].text
    temp  = soup.find('table', class_ = 'table table-striped').find_all('td')

    upc   = temp[0].text
    price_x = temp[2].text[2:]
    price_i = temp[3].text[2:]
    tax     = temp[4].text[2:]
    qn      = temp[5].text
    reviews = temp[6].text

    data.append([typ,upc,price_x,price_i,tax,qn,reviews])
```

```python
data[0]
```

```python
df2 = pd.DataFrame(data, columns=['typ', 'upc', 'price_x', 'price_i', 'tax', 'qn', 'reviews'])
```

```python
df2.head()
```

```python
df2.to_csv('book_data_part2.csv', index=False)
```

```python
df3 = pd.concat([df, df2], axis=1)
```

```python
df3.to_csv('book_data_final.csv', index=False)
```
