```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://books.toscrape.com/catalogue/page-1.html'
```

```python
res = requests.get(link)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
len(soup.find_all('li',class_= 'col-xs-6 col-sm-4 col-md-3 col-lg-3'))
```

```python
book = soup.find_all('li',class_= 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
data = []
for sp in book:
    book_name = sp.find_all('a')[-1].get('title')
    rating    = sp.find('p').get('class')[-1]
    price     = sp.find('p',class_='price_color').text[1:]
    stock     = sp.find('p',class_='instock availability').text.strip()
    book_lnk  = 'https://books.toscrape.com/catalogue/' + sp.find_all('a')[-1].get('href')
    img_lnk   = 'https://books.toscrape.com/' + sp.find('img').get('src')[2:]

    # print(sp.prettify())
    data.append([book_name, rating, price, stock, book_lnk, img_lnk])
    # print('++'*20)
    # print(f"{book_name} | {book_lnk} | {img_lnk} | {rating} | {price} | {stock}")
    # print('--'*20)
    # break
```
