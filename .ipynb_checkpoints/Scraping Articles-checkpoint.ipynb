```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://quotes.toscrape.com/')
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup
```

```python
dir(soup)
```

```python
help(soup.find)
```

```python
print(soup.find('span', class_='text').text[1:-1])
```

```python
all_span = soup.find_all('span',class_='text')
```

```python
type(all_span)
```

```python
quotes = []
for i in all_span:
    quotes.append(i.text)
```
