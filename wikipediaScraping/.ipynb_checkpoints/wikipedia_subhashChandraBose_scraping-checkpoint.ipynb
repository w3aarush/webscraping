```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
with open('subhash_chandra_sourcecode.txt','r') as fh:
    subhash_chandra_html = fh.read()
```

```python
res = requests.Response()
```

```python
res.status_code = 200
res._content = subhash_chandra_html.encode('utf-8')
res.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(res.text,'html.parser')
```

```python
heading = soup.find('h1').text
```

```python
corpus = ''
for p in soup.find_all('p'):
    corpus += (p.text)
    corpus += '\n'
    # print('==================================================\n============================================================')
corpus = corpus.strip()
```

```python
print(corpus)
```

```python
for i in range(1,387):
    corpus = corpus.replace('['+str(i)+']','')
```

```python
print(corpus)
```

```python
heading = list(heading)
```

```python
heading = ''.join(heading)
```

```python
heading = heading.replace(' ','')
```

```python
with open(heading+'.txt','w') as fh:
    fh.write(corpus)
    fh.close()
```
